## Pandas Dataframe
Python equivalent of data that is stored in a tabular format eg. CSV and SQL Database.

```python
# Creating a dataframe
# Pandas Dataframes are uilt from Python Dictionaries
# Where the key represents the column name and values are Numpy Arrays Represented as pandas series
```

```python
# 4. ACCESSING COLUMNS FROM A DATAFRAME
```

```python
data ={
    'Student Names' : ['Dorothy', 'John', 'Ann', 'Alex', 'Kelvin'],
    'Math Scores' : [89, 83, 88, 78, 90],
    'Attendance' : [True, True, False, False, True]
}
```

```python
df = pd.DataFrame(data)

# Adding a new column to the DataFrame
df['English Scores'] = [78, 83, 85, 78, 90]

# Total Marks Column
df['Total Marks'] = df['Math Scores'] + df['English Scores']
df
```

```python
# Numerical (Continuous) Column
df['Math Scores']
print(df['Math Scores'].mean())
print(df['Math Scores'].max())
print(df['Math Scores'].min())
print(df['Math Scores'].sum())
print(df['Math Scores'].count())
print(df['Math Scores'].std())
print(df['Math Scores'].var())
print(df['Math Scores'].median())
```

```python
# Textual (Categorical)
print(df['Attendance'].unique())
print(df['Attendance'].value_counts())
```

```python
# GROUPBY(): CATEGORICAL COLUMNS
df.groupby('Attendance')['Math Scores'].mean()
```

```python
df.groupby(['Attendance', 'Student Names'])['Math Scores'].mean()
```

```python
# ratio
df.groupby('Attendance').size().plot(kind='pie', autopct='%.1f%%')
```

```python
df.groupby('Attendance').size().plot(kind = 'bar')
```

```python
df.groupby('Attendance').size().plot(kind = 'line')
```
